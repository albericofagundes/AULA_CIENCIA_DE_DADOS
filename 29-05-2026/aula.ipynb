```python
from scipy.stats import norm, poisson, binom
from scipy.special import factorial
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_tentativas = 500
p_falha = 0.02
k_falhas_alvo = 10

prob = binom.pmf(k_falhas_alvo, n_tentativas, p_falha)
print(f"A probabilidade de ocorrer exatamente {k_falhas_alvo} falhas em {n_tentativas} tentativas é: {prob:.4f}")
```

```python
eventos= 0
media = 3

prob_poisson = poisson.pmf(eventos, media)
print(f"A probabilidade de ocorrer exatamente {eventos} eventos em um intervalo com média de {media} é: {prob_poisson:.4f}")
```
